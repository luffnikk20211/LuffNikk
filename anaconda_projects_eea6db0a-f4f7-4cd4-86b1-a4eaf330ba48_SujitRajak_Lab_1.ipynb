```python
# -----------------------------------------------------------
# PART 1: TAXES - Progressive Tax System (2023 rates)

# This program explains a function compute_tax() that:
# 1. Takes an income as input.
# 2. Checks which tax bracket the income belongs to.
# 3. Calculates the correct tax using formulas provided.

# -----------------------------------------------------------

def compute_tax(income):
    """
    Compute the federal tax based on 2023 U.S. tax brackets.
    Only basic if-elif conditions are used.
    """

    if income <= 11000:
        tax = 0.10 * income                           # 10% on all income up to 11,000
    elif income <= 44725:
        tax = 1100 + 0.12 * (income - 11000)          # 1,100 + 12% of amount over 11,000
    elif income <= 95375:
        tax = 5147 + 0.22 * (income - 44725)          # 5,147 + 22% of amount over 44,725
    elif income <= 182100:
        tax = 16290 + 0.24 * (income - 95375)         # 16,290 + 24% of amount over 95,375
    elif income <= 231250:
        tax = 37104 + 0.32 * (income - 182100)        # 37,104 + 32% of amount over 182,100
    elif income <= 578125:
        tax = 52832 + 0.35 * (income - 231250)        # 52,832 + 35% of amount over 231,250
    else:
        tax = 174238.25 + 0.37 * (income - 578125)    # 174,238.25 + 37% of amount over 578,125
    
    return tax

# ---- Example for part 1 ----
print("Tax on $50,000 income = $", compute_tax(50000))
print("Tax on $200,000 income = $", compute_tax(200000))
print("Tax on $1,000,000 income = $", compute_tax(1000000))


```

```python

# -----------------------------------------------------------
# PART 2: Derivative of ln(x) at x = 1 using difference formula

# Recall the definition of derivative:
#
#   f'(a) = lim (h -> 0) [ f(a+h) - f(a) ] / h
#
# For f(x) = ln(x), and a = 1:
#   f'(1) = lim (h -> 0) [ ln(1+h) - ln(1) ] / h
#
# Since ln(1) = 0, this becomes:
#   f'(1) = lim (h -> 0) ln(1+h)/h
#
# We will test this formula for smaller and smaller values of h
# (stored in a list), and watch the result approach 1.

# -----------------------------------------------------------

# We want to show that:
# (ln(1 + h) - ln(1)) / h approaches 1 as h -> 0

import math   # only basic math module

# List of h values given in the problem
h = [1, 0.1, 0.001, 0.002, 0.0001, 0.0000000001]

print("\nApproximations of derivative of ln(x) at x = 1:")

for value in h:
    numerator = math.log(1 + value) - math.log(1)     # ln(1+h) - ln(1), note ln(1)=0
    denominator = value                              # h
    derivative_approx = numerator / denominator
    print("h =", value, " => Approx derivative =", derivative_approx)

# The values get closer to 1 as h gets smaller.
```

```python

# -----------------------------------------------------------
# PART 3: Truth of Summation
# -----------------------------------------------------------

# Formula to verify:
# sum of i^3 from i=0 to n = ( (n(n+1))/2 )^2

def verify_summation(n):
    """
    Verify if the formula holds for a given n.
    """
    # Compute left-hand side (LHS): actual sum of cubes
    lhs = 0
    for i in range(n+1):     # loop from 0 to n
        lhs += i**3

    # Compute right-hand side (RHS): formula
    rhs = ((n * (n + 1)) // 2) ** 2   # integer division for safety

    return lhs == rhs   # return True if equal, False otherwise


# ---- Verify up to 1000 ----
all_true = True   # assume formula is true for all n
for n in range(1, 1001):   # check each n from 1 to 1000
    if not verify_summation(n):  # if any one fails
        all_true = False
        break

print("\nDoes the formula hold for all n ≤ 1000? ->", all_true)
```
