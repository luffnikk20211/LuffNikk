## My Favorite Function: Square Function

## I chose the square function because it is very simple and easy to understand. 

## It multiplies a number by itself and is used in many areas of math and programming, including geometry, statistics, and basic calculations.

## It is a simple function that is very versatile.




```python
#  A function is defined to square a number
#  This function takes one input 'x' and returns the result of x multiplied by itself which satisfies the square function.

def square(x):
    return x * x

# Following is a example of square of couple of numbers : 2, 5, 10

print("Square of 2:", square(2))
print("Square of 5:", square(5))
print("Square of 10:", square(10))
```

## List of Values

## I created a list of 10 random integers between 1 and 100 using Python's `random` module. 
## The `random` module is used to generate pseudo-random numbers.
## These values are different each time the code runs and demonstrate basic list operations like sum, minimum, maximum, and average.


```python
import random

# A list of 10 random numbers between 1 and 100 is generated
my_list = [random.randint(1, 100) for _ in range(10)]
print("My List:", my_list)

# calculate sum of all random list of numbers
print("Sum:", sum(my_list))

# calculate min of all random list of numbers
print("Minimum:", min(my_list))

# calculate max of all random list of numbers
print("Maximum:", max(my_list))

# calculate average of all random list of numbers
print("Average:", sum(my_list) / len(my_list))
```
